#### Valores atipicos Avg_glucose_level

```python
import numpy as np


# Seleccionar la columna que deseas analizar
columna = df_ho["Avg_glucose_level"]

# Calcular la media y la desviación estándar de la columna
media = np.mean(columna)
desv_est = np.std(columna)

# Definir el criterio para identificar valores atípicos
criterio = 2 # Si un valor está más lejos que 2 desviaciones estándar de la media, se considera atípico

# Identificar los valores atípicos
valores_atipicos = []
for num in columna:
    if abs(num - media) > criterio * desv_est:
        valores_atipicos.append(num)

print("Valores atípicos: ", valores_atipicos)
```

#### Bmi

```python
# Definir la función para clasificar las edades en grupos
def clasificar_imc(imc):
    if imc >= 0.0 and imc < 18.5:
        return "Bajo Peso"
    elif imc >= 18.5 and imc < 25.0:
        return "Peso normal"
    elif imc >= 25.0 and imc < 30.0:
        return "Sobrepeso"
    elif imc >= 30.0 and imc < 35.0:
        return "Obesidad Grado 1"
    elif imc >= 35.0 and imc < 40.0:
        return "Obesidad Grado 2"
    elif imc >= 40.0:
        return "Obesidad Grado 3"
    else:
        return "No valido"

# Crear una nueva columna "Grupo de edad" utilizando la función de clasificación
df_ho["Grupo de Imc"] = df_ho["Bmi"].apply(clasificar_imc)

# Calcular el porcentaje de pacientes en cada grupo de edad
porcentaje_bajo = (df_ho["Grupo de Imc"] == "Bajo Peso").mean() * 100
porcentaje_normal = (df_ho["Grupo de Imc"] == "Peso normal").mean() * 100
porcentaje_sobre = (df_ho["Grupo de Imc"] == "Sobrepeso").mean() * 100
porcentaje_grado1 = (df_ho["Grupo de Imc"] == "Obesidad Grado 1").mean() * 100
porcentaje_grado2 = (df_ho["Grupo de Imc"] == "Obesidad Grado 2").mean() * 100
porcentaje_grado3 = (df_ho["Grupo de Imc"] == "Obesidad Grado 3").mean() * 100
porcentaje_no_valido = (df_ho["Grupo de Imc"] == "No valido").mean() * 100

# Imprimir los resultados
print(f"Porcentaje de pacientes peso bajo: {porcentaje_bajo:.2f}%")
print(f"Porcentaje de pacientes peso normal: {porcentaje_normal:.2f}%")
print(f"Porcentaje de pacientes sobrepeso: {porcentaje_sobre:.2f}%")
print(f"Porcentaje de pacientes obesidad grado 1: {porcentaje_grado1:.2f}%")
print(f"Porcentaje de pacientes obesidad grado 2: {porcentaje_grado2:.2f}%")
print(f"Porcentaje de pacientes obesidad grado 3: {porcentaje_grado3:.2f}%")
print(f"Porcentaje de pacientes no valido: {porcentaje_no_valido:.2f}%")
```

```python
from scipy.stats import chi2_contingency

# Selecciona las columnas que son variables categóricas
categorical_columns = ['Hypertension', 'Heart_Disease', 'Married', 'Work', 'Residence', 'Smoking', 'Gender', 'Stroke']

# Crea una matriz vacía para almacenar los valores de correlación de V Cramer
v_cramer_matrix = np.zeros((len(categorical_columns), len(categorical_columns)))

# Calcula la correlación de V Cramer entre todas las combinaciones de columnas
for i, col1 in enumerate(categorical_columns):
    for j, col2 in enumerate(categorical_columns):
        contingency_table = pd.crosstab(df_clean[col1], df_clean[col2])
        chi2, _, _, _ = chi2_contingency(contingency_table)
        n = contingency_table.sum().sum()
        rows, cols = contingency_table.shape
        v_cramer = np.sqrt(chi2 / (n * (min(rows, cols) - 1)))
        v_cramer_matrix[i, j] = v_cramer

# Crea un DataFrame para mostrar los resultados
corr_table = pd.DataFrame(v_cramer_matrix, index=categorical_columns, columns=categorical_columns)

# Muestra la tabla de correlación
corr_table
```

# 2. La base de datos no está balanceada (en inglés conocida como unbalanced dataset). 
Investigue que es una base de datos no balanceada utilizando fuentes de referencia confiables (libros y papers), los problemas con este tipo de base de datos y al menos dos técnicas utilizadas durante el entrenamiento del algoritmo para reducir su efecto negativo.

Las bases de datos no balanceadas son conjuntos de datos, es decir bases de datos que no tienen un equilibrio en el número de datos correspondientes a una de las clases de interés, esto se presenta en las variables predictivas y en la variable de respuesta, sin embargo el desbalance de la base de datos es significativa y perjudicial para modelos predictivos cuanto esta se presenta en la variable de respuesta pues esta es la variable de interés y sobre la que se trabaja, comúnmente se presenta una clase minoritaria de los datos más sensibles a la experimentación, en el caso de este proyecto la clase minoritaria es de pacientes que han sufrido un ataque cerebral, el compensar el desbalanceo es esencial pues de lo contrario la aplicación de algoritmos no será igual de bueno que cuando tenemos clases balanceadas o similares, el o los algoritmos que apliquemos aprenderán mucho de una de las clases, es decir la clase mayoritaria (personas que no han tenido un ataque cerebral) y muy poco de la clase contraria, es así que el algoritmo no será un buen predictor, pues estará sesgado por su mayor conocimiento sabre la clase mayoritaria.
Por este motivo balancear los datos es muy importante en la base de datos sobre la que estamos trabajando un 94.6% de los datos son de la clase mayoritaria y un 5,4% de la clase minoritaria, existe un desbalanceo claro, existen muchos métodos para realizarlo, en este caso se aplicará un oversampling usando la técnica SMOTE y un undersampling utilizando la técnica Near-Miss (NM).

Vamos a usar un F1 Score
